#Min-Max Normalization

**We will create the column list for the 4 columns and use Min-Max Normalization.**

```python
cols = ['crim', 'zn', 'tax', 'black']
for col in cols:
    # find minimum and maximum of that column
    minimum = min(df[col])
    maximum = max(df[col])
    df[col] = (df[col] - minimum) / (maximum - minimum)
```

```python
# Create distribution plots using histplot and kdeplot
fig, axes = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
axes = axes.flatten()

for col, value in df.items():
    sns.histplot(value, ax=axes[index], kde=False)  # Histogram
    sns.kdeplot(value, ax=axes[index], color='r', linestyle='dashed')  # KDE
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```

#Input Split

Let us split the data for training and testing.

```python
# Create the feature matrix X by dropping 'medv' and 'rad' columns
X = df.drop(columns=['medv', 'rad'], axis=1)

# Create the target vector y using the 'medv' column
y = df['medv']
```

#Model Training

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
def train(model, X, y):
    # train the model
    x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42)
    model.fit(x_train, y_train)

    # predict the training set
    pred = model.predict(x_test)

    # perform cross-validation
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))

    print("Model Report")
    print("MSE:",mean_squared_error(y_test, pred))
    print('CV Score:', cv_score)
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded X and y

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Get the coefficients of the model
coef = pd.Series(model.coef_, X.columns).sort_values()

# Plot the coefficients
coef.plot(kind='bar', title='Model Coefficients')
plt.show()  # Display the plot
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import matplotlib.pyplot as plt

# Function to train the model
def train(model, X, y):
    model.fit(X, y)

# Assuming you have loaded X and y

# Create a Decision Tree Regressor model
model = DecisionTreeRegressor()

# Train the model using the train function
train(model, X, y)

# Get the feature importances of the model
importances = model.feature_importances_
coef = pd.Series(importances, X.columns).sort_values(ascending=False)

# Plot the feature importances
coef.plot(kind='bar', title='Feature Importance')
plt.show()  # Display the plot
```
